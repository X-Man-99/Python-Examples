### Overview of Examples: Positional vs Keyword Arguments, *args, and **kwargs

#### Positional vs Keyword Arguments
These examples demonstrate how Python assigns values to function parameters based on either:
- **Position** (the order of arguments)
- **Keyword** (explicitly naming each argument)

Positional arguments require the caller to follow the function’s parameter order, while keyword arguments allow flexibility by binding values to parameter names regardless of order.

#### Arbitrary Positional Arguments: `*args`
The `*args` example shows how a function can accept an arbitrary number of positional arguments.  
Inside the function, `args` becomes a **tuple**, allowing iteration, aggregation, or transformation.  
This pattern is useful when the number of inputs is unknown or variable, such as pipelines, math operations, or middleware chains.

#### Arbitrary Keyword Arguments: `**kwargs`
The `**kwargs` example demonstrates how a function can accept any number of keyword arguments.  
Inside the function, `kwargs` becomes a **dictionary**, enabling dynamic configuration, metadata handling, or flexible API design.  
This is common in frameworks where optional parameters must be captured without strict signatures.

#### Combined Use: Flexible API Modeling
The combined example shows how `*args` and `**kwargs` can coexist to model:
- Ordered data (`*args`)
- Named data (`**kwargs`)

This pattern is foundational in decorators, event systems, web frameworks, and callback handlers.

```python
def average_func(summation, num_elements = 5): # default value of num of elements is 5 unless overwritten by user at some point
    return summation/num_elements
print(average_func(15, 3))
print(average_func(15))
```

```python
# Positional v.s KeyWord args
def describe_person(name, age, occupation):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"Occupation: {occupation}")

# Positional arguments (order matters)
describe_person("Alice", 30, "Engineer")

# Keyword arguments (order does NOT matter)
describe_person(age=30, occupation="Engineer", name="Alice")
```

```python
def sum_all(*args): #args becomes a tuple and is iterable 
    total = 0
    for value in args:
        total += value
    return total

print(sum_all(1, 2, 3, 4))      # 10
print(sum_all(10, 20))         # 30
```

```python
def debug_settings(**kwargs): #KWARGS becomes a dictionary with the key word well as a KEy in the dictionary
    for key, value in kwargs.items():
        print(f"{key} = {value}")

debug_settings(mode="test", verbose=True, retries=3)
```

```python
def model(*args, **kwargs):
    # enforce that args must be numbers
    for a in args:
        if not isinstance(a, (int, float)):
            raise TypeError("All positional args must be numeric")

    # enforce that kwargs must be strings
    for key, value in kwargs.items():
        if not isinstance(value, str):
            raise TypeError("All keyword values must be strings")

    return args, kwargs
```
