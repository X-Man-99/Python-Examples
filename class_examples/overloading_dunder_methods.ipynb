# Overview
We saw an example of this in the customer iterators class file. This will explore other dunder methods notably the __eq__() comparison dunder method and __str__(). Where we can define how to compare two object types

```python
class Employee:
    MIN_SALARY = 30000   # class-level minimum salary - no role beneath this. Class attribute used to check and set a global minimum for data in object instances

    def __init__(self, name, role, salary):
        self.name = name
        self.role = role
        self.salary = max(salary, Employee.MIN_SALARY)

    def update_role(self, new_role):
        self.role = new_role
        print(f"{self.name}'s role updated to {self.role}")

    def update_salary(self, new_salary):
        if new_salary < Employee.MIN_SALARY:
            raise ValueError(f"Salary cannot be below minimum: {Employee.MIN_SALARY}")
        self.salary = new_salary
        print(f"{self.name}'s salary updated to {self.salary}")

    def give_raise(self, amount):
        new_salary = self.salary + amount
        if new_salary < Employee.MIN_SALARY:
            raise ValueError("Raise results in salary below minimum.")
        self.salary = new_salary
        print(f"{self.name} received a raise. New salary: {self.salary}")
    def __str__(self):
        return f"{self.name} has the role of {self.role} and earns ${self.salary}"
    def __eq__(self, other):
        if not isinstance(other, Employee):
            return NotImplemented
        return (self.name,self.role) == (other.name, other.role)
```

```python
class Manager(Employee):
    MIN_SALARY = 50000

    def __init__(self, name, salary, department):
        super().__init__(name, "Manager", salary) # calls previous class's (employee) instructor
        self.department = department

    def __str__(self):
        return f"{self.name} (Manager of {self.department}) — ${self.salary}"

    def __eq__(self, other):
        if not isinstance(other, Manager):
            return NotImplemented
        return self.department == other.department
```

```python
e1 = Employee("Alice", "Developer", 270000)
e2 = Employee("Alice", "Developer", 70000)
m1 = Manager("Bob", 90000, "Engineering")
m2 = Manager("Jim", 190000, "Engineering")
```

```python
print(e1)   # Print calls on the __str__() method within the class which returns the format to print
print(m1)   # Called by print(obj) or str(obj)
```

# Notes on __eq__()
The default __eq__ behavior in Python compares object identity — meaning two objects are considered equal only if they are literally the same object in memory.

```python
# __eq__ compares meaningful fields - Avoids default identity comparison, which is almost never what you want for domain objects like employees or managers 
print(e1 == e2)   # True
print(m1 == m2)   # True
print(e1 == m1)   # False (different types → NotImplemented → False)
```

```python
class NoEqMethodClass:
    def __init__(self,name ,role):
        self.name = name
        self.role = role
        self.salary = 270000
no_eq_instance = NoEqMethodClass("Alice", "Developer")
other_no_eq_instance = NoEqMethodClass("Alice", "Developer")
different_no_eq_instance = NoEqMethodClass("Dallas", "Developer")
print(no_eq_instance == other_no_eq_instance) # Despite the exact same values, these instances are at two different memory locations and therefore not equal
print(no_eq_instance == different_no_eq_instance) # Diff. vinstance with diff values
print(no_eq_instance == no_eq_instance) # literally compares the instance to itself, no shit its equal
```
