# Overview
Examples of basic iterables, identify iterable objects. The Next() function.
Files are also iterable and commonly used with generators to stream chunks of data instead of all at once
Apply Iter( function to iterable objects that can be converted into iterables)

```python
import pandas as pd
list_of_num = [1,2,3,4,5,6,7]
string = "Hello World" 
dictionary = {"name": "John", "age": 30, "city": "New York"}
data = {'Name': ['Tom', 'Nick', 'Krish', 'Jack'], 'Age': [20, 21, 19, 18]}
df = pd.DataFrame(data)

print(df)
```

```python
it_list = iter(list_of_num)
it_string = iter(string)
it_dict = iter(dictionary)
it_df = iter(df)

iterable = [it_list, it_string, it_dict, it_df]
```

```python
print("next list element:", next(it_list))
print("===================================")
print("next string element:", next(it_string)) #outputs a single character each time run until the end of the string is reached at which point it will generate out of range error
print("===================================")
print("next dictionary element:", next(it_dict )) #notice it iterates over the keys of the dictionary, not the values
print("===================================")
print("next DataFrame element:", next(it_df)) # notice it iterates over the index of the DataFrame, not the rows

print("===================================")
for i in iterable:
    print("next element:", next(i)) #outputs the next element in each iterabled
print("===================================")
print(iterable) #outputs the list of iterables mem locations
```
