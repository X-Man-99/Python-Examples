```python
class Employee:
    MIN_SALARY = 30000   # class-level minimum salary - no role beneath this. Class attribute used to check and set a global minimum for data in object instances

    def __init__(self, name, role, salary):
        self.name = name
        self.role = role
        self.salary = max(salary, Employee.MIN_SALARY)

    def update_role(self, new_role):
        self.role = new_role
        print(f"{self.name}'s role updated to {self.role}")

    def update_salary(self, new_salary):
        if new_salary < Employee.MIN_SALARY:
            raise ValueError(f"Salary cannot be below minimum: {Employee.MIN_SALARY}")
        self.salary = new_salary
        print(f"{self.name}'s salary updated to {self.salary}")

    def give_raise(self, amount):
        new_salary = self.salary + amount
        if new_salary < Employee.MIN_SALARY:
            raise ValueError("Raise results in salary below minimum.")
        self.salary = new_salary
        print(f"{self.name} received a raise. New salary: {self.salary}")
```
