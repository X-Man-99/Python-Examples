# Overview
class methods that are consistent across all instances of a class. Typical for default values across all object instance.
Class methods should only access class level attributes / data. Handling the high variance in instances isn't feasible like the stability of class level attributes.
***Instance Level Attributes*** - Differ among instances
***Class level Attributes*** - Great for security as not accessible via typical methods, usually used to enable alternative constructors, returns class object instead of using constructor to build an instance.

```python
import random
```

```python
class MyClass():
    @classmethod
    def cls_method_id_generation(cls,seed):
        cls.id = random.random()* seed
        print(cls.id)
    def __init__(self):
        self.string = 'init string'
    def my_method(self, custom_input):
        self.string = custom_input
        print('pass' , self.string,)
    def your_method(self):
        self.id = MyClass.id # notice this ID belongs to the class so we call the class not an instance to access it
        print('you pass',self.id , self.string)
```

```python
my_object = MyClass()
my_object.cls_method_id_generation(2445)
my_object.my_method("HELLO")
my_object.your_method()
```

```python
print("class ID:",MyClass.id)
```

```python
print('Notice two objects of same class will have diff self.string values but same Class.id: ')
second_object = MyClass()
print(my_object.your_method(),second_object.your_method())
print('=======================================================================================================')
print('Notice same MyClass.id Values but different string slighty because of the redefinition of the instance attribute self.string in the first instance')
```
