```python
import random
# initialize 10 random integers from 1 - 20
count = 10
start = 1
end = 20
random_list = [random.randint(start, end) for element in range(count)]

print("Random list of ten values 1- 20: ", random_list)



```

```python
# print only numbers greater than 10
print("Numbers greater than 10: ", [element for element in random_list if element > 10])
```

```python
# print all numbers in the list but only square the even numbers
print(
    "Square of even numbers: ", 
    [ element**2 if element % 2 == 0 else element for element in random_list]
)
```

# Summary
## Similarities
- Both comprehensions are encapsulated by '[]' square brackets ensuring that the operation that occurs within is output finally as a list
- For loop operates on the entire list with same bounds
## Differences
- Logic
  - The if statement at the end of the "Numbers Greater than 10" expression acts as a filter
    -  A filter that decides whether the expression is evaluated and added to the result list.
  - The loop will add all elements to the resulting list with no filter but the logic at the beginning determines if additional manipulation in this case squaring occurs
