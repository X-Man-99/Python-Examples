```python
# create dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5,  'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10}
```

```python
#outputs list of tuples of key-value pairs
items = my_dict.items()
print("Items of the dictionary:", type(items), items)
```

```python
# comprehension to create a new dictionary with keys and values swapped
swapped_dict = {value: key for key, value in my_dict.items()}
print("Swapped dictionary:", type(swapped_dict), swapped_dict)

scaled_dict = {key: value * 10 for key, value in my_dict.items()}
print("Scaled dictionary:", type(scaled_dict), scaled_dict)
```

```python
# Creating a Dictionary from Two Lists
keys = ['a','b','c','d','e']
values = [1, 2, 3, 4, 5]  

d = {k:v for (k,v) in zip(keys, values)}  
print (d)
```
