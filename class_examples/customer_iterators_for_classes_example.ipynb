# Overview
***Dunder Methods:*** - dunder methods are Python’s built‑in “default functions” that define how an object behaves.  
But the deeper truth is even better: they are the hooks that let your objects participate in Python’s syntax itself. Default behaviors attached to every object type, which you can override.
For something to be an iterator object it must utilize these two dunder methods ***__iter__()*** - returns self as an iterator object & ***__next__()*** - returns the next value of an iterator.

Both of these dunder methods must be defined in a class for it to be iterable

These do not behave like list that will eventually throw an out of bounds exception. Must define ***StopIteration as stop condition in the __next__() method***

```python
class MyIteratorClass:
    def __inti__(self, start):
        self.counter = start
    def __iter__(self):
        return self # as defined above this returns self as an iterator
    def __next__(self):
        if self.counter <= 0:
            raise StopIteration('NO MORE NEXT!')
        print('Current Count: ', self.counter)
        self.counter -= 1
        return self.counter
```
