```python
# Therefore we create a function that will contain the generator and will yield the values one by one. This will stream data in chunks instead of loading the entire file into memory at once. This is especially useful for large files or video streaming
def num_sequence_generator(n):
    i = 0
    while i < n:
        yield i #yield keyword defines or what makes the generator function. It is responisble for the creation of generator object.
        i+= 1
```

```python
result = num_sequence_generator(5)
print(type(result)) # notice of generator object being created. This is a lazy evaluation, meaning that the values are not generated until they are requested. As opposed to eager evaluation, meaning that all the values are generated at once and stored in memory. This can be inefficient for large datasets or infinite sequences.
for num in result:
    print(num)
```

```python
# now with files and pandas, we can use the generator to read the file in chunks and process it in a memory-efficient way. This is especially useful for large files that cannot fit into memory all at once.
def read_file_in_chunks(file_path): # chunk size = num of lines to read at a time
    with open(file_path, 'r') as file:
        it_file = iter(file)
        while True: # if no loop it will return StopIteration error. So we need to use a while loop to keep yielding the next line until the end of the file is reached.
            yield next(it_file) #can use pd.read_csv as well that returns an iterable object that can be used to read the file in chunks.
```
