# Overview
While this notebook uses dictionaries please note that list_comprehension v.s dictionary_comprehension is defined by what datatype is returned.

```
return [num for num in number_list]
```
v.s
```
return {key , num for key, num in number_dict}
```
Therefore, returning a list '[]' of keys despite being on operation on a dictionary is output as a list

````python
# create dictionary
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5,  'f': 6, 'g': 7, 'h': 8, 'i': 9, 'j': 10}
````

````python
# Notice that the keys, values, and items are not lists.
keys = my_dict.keys()
print("Keys of the dictionary:", type(keys), keys)

values = my_dict.values()
print("Values of the dictionary:", type(values), values)

items = my_dict.items()
print("Items of the dictionary:", type(items), items)
````

````python
# can convert to list several ways
keys_list = list(keys)
values_list = list(values)
items_list = list(items)

# comprehension to create a list of keys, values, and items
keys_list_comp = [key for key in my_dict.keys()]
values_list_comp = [value for value in my_dict.values()]
items_list_comp = [item for item in my_dict.items()]

print("Values of the dictionary as list:", type(values_list), values_list)
print("Values of the dictionary as list via comprehension:", type(values_list_comp), values_list_comp)
print("Items of the dictionary as list via comprehension:", type(items_list_comp), items_list_comp)
print("Items of the dictionary as list:", type(items_list), items_list) # list of tuples
print("Datatype of items in the list of items:", type(items_list[0]))
````
