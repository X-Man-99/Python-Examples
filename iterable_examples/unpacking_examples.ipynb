# Overview
Examples describing common scenarios for unpacking with the '*' operator and various syntax in which it may be used.
Unpacking lets you assign multiple variables at once.

```python
# basic unpacking
a, b = [1, 2]
x, y, z = (10, 20, 30)
print("a = {}, b = {}".format(a, b))
print("x = {}, y = {}, z = {}".format(x, y, z))
# unpacking with *
first, *middle, last = [1, 2, 3, 4, 5]
print("first = {}, middle = {}, last = {}".format(first, middle, last))
```

# How it works
Python evaluates the right side first, producing a sequence. Then it assigns each element to the corresponding variable on the left.

```python
# unpacking function arguments
def my_function(a, b, c):
    print("a = {}, b = {}, c = {}".format(a, b, c))

my_list = [1, 2, 3]
my_function(*my_list)  # Unpacking the list into function arguments
```

# How it works
The starred variable (middle) becomes a list containing all items not matched by other variables.

# Why it works
Python allows exactly one starred target in an assignment.
It expands the iterable, assigns fixed positions first, and dumps the remainder into the starred variable.
This is Python’s way of supporting “variadic” unpacking at assignment time.

```python
def greet(name, age):
    print(f"{name} is {age} years old")

info = {"name": "Alice", "age": 30}
greet(*info)
greet(**info)  # Unpacking the dictionary into function arguments
```

Double‑unpacking (**) is Python’s mechanism for expanding a mapping (usually a dict) into keyword arguments or into another dictionary literal.  
It’s the “dictionary version” of the * operator.
info - is the baseline dictionary data type
*info - is the keys unpacked
**info - it the values in the dictionary


```python
# unpacking in loops
for a, b in [(1, 2), (3, 4), (5, 6)]: 
    print("a = {}, b = {}".format(a, b))
```

```python
#unpacking in comprehensions
pairs = [(1, 2), (3, 4), (5, 6)]
sums = [a + b for a, b in pairs]
print(sums)
```

```python
string = "Hello"
iterable = iter(string)
print(next(iterable))  # H
print(next(iterable))  # e  
print(*iterable)  # l l o
```

# How unpacking works in this context
1. iter(string) creates an iterator
iter("Hello") produces an iterator object that yields characters one at a time.

Internally, it keeps a cursor (or pointer) to the current position.

2. next(iterable) advances the iterator
Each call to next() moves the cursor forward:

First next() → "H"

Second next() → "e"

Now the iterator’s internal pointer is positioned at "l".

3. print(*iterable) consumes the rest of the iterator
The * operator asks Python to:

Iterate through the remaining items and expand them into separate arguments.

# Why you cannot unpack it again
Iterators in Python are single‑pass.
Once an iterator yields an item, that item is gone forever.

There is no rewinding, no resetting, no second pass.

Why?
Because an iterator is defined by two rules:

It must return items one at a time.

It must raise StopIteration when it has no more items.

After print(*iterable) finishes, the iterator has reached the end of the underlying sequence. Its internal pointer is now out of bounds.
