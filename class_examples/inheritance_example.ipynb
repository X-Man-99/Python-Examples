# Overview
Inheritance in Python allows a class to reuse and extend the attributes and methods of another class. This supports cleaner design, reduces duplication, and enables objects to build on previously defined behaviors. Multiple inheritance expands this idea further by allowing a class to inherit from more than one parent class, combining capabilities, permissions, or interfaces into a single object.

With inheritance, a class can:
- automatically gain attributes and methods from its parent
- override or extend inherited behavior
- layer additional functionality on top of existing “permissions” or capabilities
- combine behaviors from multiple sources using Python’s method resolution order (MRO)

This makes it possible to construct complex systems where classes share common logic, specialize behavior, or merge features from multiple base classes.

````python
class Employee:
    MIN_SALARY = 30000   # class-level minimum salary - no role beneath this. Class attribute used to check and set a global minimum for data in object instances

    def __init__(self, name, role, salary):
        super().__init__()
        self.name = name
        self.role = role
        self.salary = max(salary, Employee.MIN_SALARY)

    def update_role(self, new_role):
        self.role = new_role
        print(f"{self.name}'s role updated to {self.role}")

    def update_salary(self, new_salary):
        if new_salary < Employee.MIN_SALARY:
            raise ValueError(f"Salary cannot be below minimum: {Employee.MIN_SALARY}")
        self.salary = new_salary
        print(f"{self.name}'s salary updated to {self.salary}")

    def give_raise(self, amount):
        new_salary = self.salary + amount
        if new_salary < Employee.MIN_SALARY:
            raise ValueError("Raise results in salary below minimum.")
        self.salary = new_salary
        print(f"{self.name} received a raise. New salary: {self.salary}")
````

````python
class Manager(Employee):
    MIN_SALARY = 50000

    def __init__(self, name, salary, department):
        super().__init__(name, "Manager", salary) # calls previous class's (employee) instructor
        self.department = department
````

````python
m1 = Manager('Alice', 50000, 'Engineering')
m1.give_raise(10000) # sEE THAT THE  INSTANCE OF THE MANAGER CLASS HAS THE PREVIOUS METHODS OF THE EMPLOYEE CLASS ON IT WITHOUT BEING EXPLICITLY DEFINED IN ITS BODY
````

# MultiInheritance
With multiple inheritance, super() does NOT call both constructors automatically unless the parent classes cooperate.
```super.__init()``` Does not "simply call the parents" it only invokes the NEXT __init__() / class in the MRO. Therefore in the case of intern it only calls student and the student constructor then needs to call employee with its own super.
Even if Student does not inherit from Employee, it still participates in the MRO chain when used. “Call the next class in the MRO after Student." And in this case, the next class is Employee,
## Which level of inheritance resolves first what is overriden if there is overlap?
***Method Resolution Order (MRO)*** - MRO is the order Python uses to search for attributes and methods when you access them on an object. 
- Intern → Student → Employee → object

- Intern.__init__ runs first
- If it calls super(), Python goes to Student.__init__
- If Student.__init__ calls super(), Python goes to Employee.__init__
- If Employee.__init__ calls super(), Python goes to object.__init__

Attribute lookup follows MRO, If both parents define the same attribute, the employee.name field is prioritized as it appears first in MRO or literally further to the left in the class definition
``` class Intern(EMPLOYEE, then Student)``` Employee has priority and will override student.name

````python
class Student:
    def __init__(self):
        super().__init__()
        self.name = "billy the kid"
        self.classification = "freshman"
        self.gpa = 4.0
````

````python
class Intern(Employee, Student): # Order here matters and what takes priority in MRO should be conscious decision effects what data is needed upon instantiation if Employee needs data but student is first and doesn't pass the data along from the constructor of the instantiated intern class consider switching which comes first in MRO and consider which one should overwrite the other
    def __init__(self, name, role, salary):  
        super().__init__(name,role,salary)
````

````python
intern = Intern("Mick jenkins",'Engineer',30000) # notice students hardcoded values dont come through because of position in MRO
print (intern.name)
print(intern.gpa)
print(intern.role)
````

````python
class Intern(Employee, Student): # You can manually call both constructors:
    def __init__(self, name, role, salary):    
        Student.__init__(self)
        Employee.__init__(self, name, role, salary)
````

````python
intern2 = Intern("Mick jenkins",'Engineer',30000)
print(intern2.name)
print(intern2.gpa)  # Has Student Attributes
print(intern2.role) # Has Employee Attributes
````
