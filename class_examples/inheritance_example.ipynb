# Overview
Inheritance in Python allows a class to reuse and extend the attributes and methods of another class. This supports cleaner design, reduces duplication, and enables objects to build on previously defined behaviors. Multiple inheritance expands this idea further by allowing a class to inherit from more than one parent class, combining capabilities, permissions, or interfaces into a single object.

With inheritance, a class can:
- automatically gain attributes and methods from its parent
- override or extend inherited behavior
- layer additional functionality on top of existing “permissions” or capabilities
- combine behaviors from multiple sources using Python’s method resolution order (MRO)

This makes it possible to construct complex systems where classes share common logic, specialize behavior, or merge features from multiple base classes.
