# Overview
attribute access customization in Python using __getattr__, __setattr__, and related hooks. These dunder methods are part of Python’s data model and give you fine‑grained control over how attributes are retrieved, assigned, validated, or synthesized.
Python normally handles attribute access through its built‑in mechanisms:
- obj.x → direct attribute lookup
- obj.x = value → direct assignment
- missing attributes → AttributeError
But sometimes you need more control:
- computed or dynamic attributes
- validation on assignment
- logging or debugging
- proxy objects
- enforcing read‑only or write‑only fields
- fallback behavior when attributes don’t exist

```python
class Employee:
    MIN_SALARY = 30000

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __setattr__(self, name, value):
        if name == "salary" and value < Employee.MIN_SALARY:
            raise ValueError("Salary below minimum")
        super().__setattr__(name, value)

    def __getattr__(self, name):
        if name == "email":
            return f"{self.name.lower()}@company.com"
        raise AttributeError(f"{name} not found")
```

```python
e = Employee("Alice", 50000)
print(e.email)     # auto-generated
e.salary = 10000   # ValueError
```
