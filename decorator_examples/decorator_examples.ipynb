# Overview
Function decorators are wrappers to augment other functions we will explore syntax and explain this wrapper behavior.

## Use cases for decorators
You use decorators when you want to add behavior to functions in a clean, reusable, centralized way — without modifying the functions themselves.
Repeated functionality like logging or augmenting data from data source before passing to other functions that need the data.
Use case:
- type checking
- argument validation
- authentication (Flask, FastAPI)
- retry logic
- caching
- Logging
- modify arguments
- modify return values
- run code before the function
- run code after the function
- replace the function entirely
- Decorators register callbacks and pipelines.
- decorator registers the function as a web route -web API
- To separate concerns cleanly 
  - Decorators let you keep your core logic pure.


```python
def decorator_double(func):
    def wrapper (*args,**kwargs):
        doubled_args = [a * 2 for a in args]
        return func(*doubled_args,**kwargs)
    return wrapper
```

```python
@decorator_double
def multiply_values(a,b):
    return a*b
print(multiply_values(1,5)) # double arguments and quadruples output to 20
```

# Explanation
Python interprets this as multiply_values = decorator_double(multiply_values)
## But where are the parenthesis for multiply values and saving functions to variables
The decorator is nothing more than a function that returns a new function, and Python assigns that new function to the original name.

Because Python treats functions as first‑class objects:

- You can store them in variables
- You can pass them into other functions
- You can return them from functions
- You can replace them dynamically

Decorators exploit this last ability:
They replace the original function with a modified version.

```python
func_as_var = multiply_values
print(func_as_var(20, 2)) # rename the function 

#DECORATORS CAN BE MANUALLY INVOKED like this
final_decorator_quadrupled = decorator_double(func_as_var)
print('==============================\n', final_decorator_quadrupled(2,20))
```
